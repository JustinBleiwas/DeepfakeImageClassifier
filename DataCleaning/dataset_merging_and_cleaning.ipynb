```python
# Merge datasets and rename files

import os
from PIL import Image

# Paths to your cleaned datasets
cleaned_datasets = [
    "dataset_1_cleaned",
    "dataset_2_cleaned",
    "dataset_3_cleaned"
]

output_dir = "merged_dataset"
real_dir = os.path.join(output_dir, "real")
fake_dir = os.path.join(output_dir, "fake")

os.makedirs(real_dir, exist_ok=True)
os.makedirs(fake_dir, exist_ok=True)

real_count = 1
fake_count = 1

for dataset in cleaned_datasets:
    print(f"Merging: {dataset}")

    for cls in ["real", "fake"]:
        src_dir = os.path.join(dataset, cls)
        if not os.path.isdir(src_dir):
            continue

        for fname in os.listdir(src_dir):
            src_path = os.path.join(src_dir, fname)

            # Load and ensure RGB JPG
            try:
                img = Image.open(src_path).convert("RGB")
            except:
                print("Warning: could not open", src_path)
                continue

            # Determine output path
            if cls == "real":
                out_name = f"real_{real_count}.jpg"
                out_path = os.path.join(real_dir, out_name)
                real_count += 1
            else:
                out_name = f"fake_{fake_count}.jpg"
                out_path = os.path.join(fake_dir, out_name)
                fake_count += 1

            # Save
            img.save(out_path, "JPEG", quality=95, subsampling=2)

print("Done merging!")
print(f"Total real images: {real_count - 1}")
print(f"Total fake images: {fake_count - 1}")
```
